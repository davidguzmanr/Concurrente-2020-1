## Ejercicio 1 (productor-consumidor)

Lo siguiente es lo que expuso Dante del problema del productor-consumidor usando Mutex.

```python
from threading import Thread, BoundedSemaphore, Lock
lock = Lock()
from queue import Queue

class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.lock = Lock()
    
    def bloquear(self):
        self.lock.acquire()
        self.esta_bloqueado = True
    
    def desbloquear(self):
        self.lock.release()
        self.esta_bloqueado = False
```

```python
class Productor(Thread):
    def __init__(self,buffer,mutex):
        super().__init__()
        self.buffer = buffer
        self.mutex = mutex
        self.lock = Lock()
        self.esta_despierto = False
        
    def producir(self):
        # Sólo se agrega uno
        item = 1
        print('Productor: Acabo de producir un item')
        return item
    
    def añadir_al_buffer(self,item):
        self.buffer.put(item)
        print('Productor: Agregué un item al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer.full():
                    print('Productor: Hay espacio para nuevos items')
                    item = self.producir()
                    self.añadir_al_buffer(item)
                self.mutex.desbloquear()
```

```python
class Consumidor(Thread):
    def __init__(self,buffer,mutex):
        super().__init__()
        self.buffer = buffer
        self.mutex = mutex
        self.lock = Lock()
        self.esta_despierto = False
    
    def quitar_del_buffer(self):
        item = self.buffer.get(1)
        print('Consumidor: Obtuve un item del buffer')
        return item
    
    def consumir(self,item):
        print('Consumidor: Consumí un item')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer.empty():
                    print('Consumidor: Hay items por consumir')                    
                    item = self.quitar_del_buffer()
                    self.consumir(item)
                self.mutex.desbloquear()
```

Hacemos un productor y 2 consumidores.

```python
mutex = Mutex()
buffer = Queue(50)

productor = Productor(buffer,mutex)
consumidor1 = Consumidor(buffer,mutex)
consumidor2 = Consumidor(buffer,mutex)

productor.start()
consumidor1.start()
consumidor2.start()

productor.join()
consumidor1.join()
consumidor2.join()
```

## Ejercicio 2 (composición de funciones)

```python
import numpy as np
import matplotlib.pyplot as plt
import threading
import time
from threading import Thread, BoundedSemaphore, Lock
lock = Lock()
from queue import Queue

class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.lock = Lock()
    
    def bloquear(self):
        self.lock.acquire()
        self.esta_bloqueado = True
    
    def desbloquear(self):
        self.lock.release()
        self.esta_bloqueado = False
```

```python
def g(x):
    return 3*x**2 + 2*x -1

def f(y):
    return np.sin(y) + np.cos(y)
```

```python
class Composicion(threading.Thread):
    
    def __init__(self, funcion, x, array, mutex):
        super(Composicion, self).__init__()
        self.funcion = funcion
        self.x = x
        self.array = array
        self.mutex = mutex
        
    def run(self):
        
        global x
        
        self.mutex.bloquear()
        
        print(self.funcion.__name__ + "(" + str(x) + ")" + " = " + str(self.funcion(self.x)) + "\n")
        x = self.funcion(x)
        self.array.append(x)
        
        self.mutex.desbloquear()
            
```

```python
xs = np.arange(-100, 100.25, 0.25)

ys_g = []
ys_f = []

mutex = Mutex()

for i in range(len(xs)):
    
    x = xs[i]
    
    thread_g = Composicion(g, x, ys_g, mutex)
    thread_f = Composicion(f, x, ys_f, mutex)
        
    thread_g.start()
    thread_f.start()
    
    thread_g.join()
    thread_f.join()
    
```

```python
plt.plot(xs, ys_g)
plt.title('g(x) = 3x^2 + x^2 - 1')
plt.grid()
plt.show()

plt.plot(xs, ys_f)
plt.title('f(x) = sin(g(x)) + cos(g(x))')
plt.grid()
plt.show()
```

En lo anterior vemos que se hace la composición de funciones $f(g(x))$:

\begin{align}
\text{primer hilo}&: \quad g(x) = y\\
\text{segundo hilo}&: \quad f(y) = z
\end{align}

para $x \in [-100, 100]$ en intervalos de $\Delta x = 0.25$.

## Ejercicio 3 (máquinas de chicle)

Necesitamos un buffer para la maquina productora de chicles (buffer_chicle) y otro buffer para la maquina productora de polvo para chicles (buffer_polvo), el buffer lo podemos implementar como una cola. La máquina que recibe el polvo y el chicle toma una unidad de polvo y chicle del buffer correspondiente y pone el bolichicle (chicle cubierto con el polvo) en buffer_bolichicle. La máquina que envuelve los bolichicles toma un bolichicle de buffer_bolichicle y lo envuelve, para luego ponerlo en buffer_envueltos. Finalmente, la última máquina toma 10 bolichicles envueltos de buffer_envueltos y los pone en una bolsa. En cada parte debemos tener cuidado de que el buffer de donde se toman elementos no esté vacío y que tampoco se pongan más elementos en el buffer correspondiente si está lleno.

```python
from threading import Thread, BoundedSemaphore, Lock
lock = Lock()
from queue import Queue
```

```python
class productor_chicle(Thread):
    def __init__(self,buffer_chicle,mutex):
        super().__init__()
        self.buffer_chicle = buffer_chicle
        self.mutex = mutex
        self.lock = Lock()
        
    def producir_chicle(self):
        # Añade sólo uno al buffer
        item_chicle = 1
        print('Productor chicle: Acabo de producir un chicle')
        return item_chicle
    
    def añadir_al_buffer_chicle(self,item_chicle):
        self.buffer_chicle.put(item_chicle)
        print('Productor chicle: Agregué un chicle al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer_chicle.full():
                    print('Productor chicles: Hay espacio para chicles')
                    item_chicle = self.producir_chicle()
                    self.añadir_al_buffer_chicle(item_chicle)
                self.mutex.desbloquear()
```

```python
class productor_polvo(Thread):
    def __init__(self,buffer_polvo,mutex):
        super().__init__()
        self.buffer_polvo = buffer_polvo
        self.mutex = mutex
        self.lock = Lock()
        
    def producir_polvo(self):
        # Añade sólo uno al buffer
        item_polvo = 1
        print('Productor polvo: Acabo de producir un polvo')
        return item_polvo
    
    def añadir_al_buffer_polvo(self,item_polvo):
        self.buffer_polvo.put(item_polvo)
        print('Productor polvo: Agregué un polvo al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer_polvo.full():
                    print('Productor polvo: Hay espacio para polvo')
                    item_polvo = self.producir_polvo()
                    self.añadir_al_buffer_polvo(item_polvo)
                self.mutex.desbloquear()
                    ##
```

```python
class maquina_bolichicle(Thread):
    def __init__(self,buffer_chicle, buffer_polvo, buffer_bolichicle, mutex):
        super().__init__()
        self.buffer_chicle = buffer_chicle
        self.buffer_polvo = buffer_polvo
        self.buffer_bolichicle = buffer_bolichicle
        self.mutex = mutex
        self.lock = Lock()
    
    def quitar_del_buffer_chicle(self):
        item = self.buffer_chicle.get(1)
        print('Máquina cubre chicles: Tomé un chicle')
        return item
    
    def quitar_del_buffer_polvo(self):
        item = self.buffer_polvo.get(1)
        print('Máquina cubre chicles: Tomé un polvo')
        return item
    
    def hacer_bolichicle(self):
        # Hace sólo una a la vez
        item = 1
        print('Máquina cubre chicles: Hice un bolichicle')
        return item
        
    def añadir_al_buffer(self, item):
        self.buffer_bolichicle.put(item)
        print('Máquina cubre chicles: Agregué un bolichicle al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer_bolichicle.full() and not self.buffer_chicle.empty() and not self.buffer_polvo.empty():
                    print('Máquina bolichicle: Hay chicle y polvo disponible')
                    print('Máquina bolichicle: Hay espacio para bolichicle')
                    item_chicle = self.quitar_del_buffer_chicle()
                    item_polvo = self.quitar_del_buffer_polvo()
                    item_bolichicle = self.hacer_bolichicle()
                    self.añadir_al_buffer(item_bolichicle)
                self.mutex.desbloquear()
                    
```

```python
class maquina_envuelve_chicle(Thread):
    def __init__(self,buffer_bolichicle, buffer_envuelto, mutex):
        super().__init__()
        self.buffer_bolichicle = buffer_bolichicle
        self.buffer_envuelto = buffer_envuelto
        self.mutex = mutex
        self.lock = Lock()
        
    def quitar_del_buffer_bolichicle(self):
        item_bolichicle = self.buffer_bolichicle.get(1)
        print('Máquina envuelve chicles: Tomé un bolichicle.')
        return item_bolichicle        
        
    def envolver_bolichicle(self):
        item_envuelto = 1
        print('Máquina envuelve chicles: Envolví un bolichicle.')
        return item_envuelto 
    
    def añadir_al_buffer_envuelto(self,item_envuelto):
        self.buffer_envuelto.put(item_envuelto)
        print('Máquina envuelve chicles: Agregué un bolichicle envuelto al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer_envuelto.full() and not self.buffer_bolichicle.empty():
                    print('Máquina envuelve chicles: Hay bolichicles por envolver')
                    self.quitar_del_buffer_bolichicle()
                    item_envuelto  = self.envolver_bolichicle()
                    self.añadir_al_buffer_envuelto(item_envuelto)
                self.mutex.desbloquear()
```

```python
class maquina_embolsa_chicles(Thread):
    def __init__(self, buffer_envuelto, buffer_embolsados, mutex):
        super().__init__()
        self.buffer_envuelto = buffer_envuelto
        self.buffer_embolsados = buffer_embolsados
        self.mutex = mutex
        self.lock = Lock()
        
    def quitar_del_buffer_envuelto(self):
        item_envuelto = self.buffer_envuelto.get(10)
        print('Máquina embolsa chicles: Tomé 10 bolichicles envueltos.')
        return item_envuelto        
        
    def embolsar_chicles(self):
        item_embolsado = 1
        print('Máquina embolsa chicles: Puse 10 bolichicles en una bolsa.')
        return item_embolsado
    
    def añadir_al_buffer_embolsado(self,item_embolsado):
        self.buffer_envuelto.put(item_embolsado)
        print('Máquina embolsa chicles: Agregué una bolsa de chicles al buffer')
    
    def run(self):
        while True:
                self.mutex.bloquear()
                if not self.buffer_embolsados.full() and not self.buffer_envuelto.empty():
                    print('Máquina embolsa chicles: Hay chicles por embolsar')
                    self.quitar_del_buffer_envuelto()
                    item_embolsado  = self.embolsar_chicles()
                    self.añadir_al_buffer_embolsado(item_embolsado)
                # Como no hay nada que quite las bolsas si el buffer con bolsas se llena 
                # se las quitamos manualmente para que el proceso continue
                elif self.buffer_embolsados.full():
                    self.buffer_embolsados.get(5)
                self.mutex.desbloquear()
```

```python
# No parece haber diferencia si le ponemos el mismo mutex a los 5 hilos o mutex diferentes a cada uno.

mutex1 = Mutex()
mutex2 = Mutex()
mutex3 = Mutex()
mutex4 = Mutex()
mutex5 = Mutex()

buffer_chicle = Queue(50) 
buffer_polvo = Queue(50) 
buffer_bolichicle = Queue(50) 
buffer_envuelto = Queue(50) 
buffer_embolsados = Queue(50)


chicle = productor_chicle(buffer_chicle,mutex1)
polvo = productor_polvo(buffer_polvo, mutex2)
bolichicle = maquina_bolichicle(buffer_chicle, buffer_polvo, buffer_bolichicle, mutex3)
envolver = maquina_envuelve_chicle(buffer_bolichicle, buffer_envuelto, mutex4)
embolsar = maquina_embolsa_chicles(buffer_envuelto, buffer_embolsados, mutex5)

chicle.start()
polvo.start()
bolichicle.start()
envolver.start()
embolsar.start()

chicle.join()
polvo.join()
bolichicle.join()
envolver.join()
embolsar.join()
```
